```python
import pandas as pd

import json

from datasets import load_dataset
 
# ---------- STEP 1: Load and Convert Hugging Face Dataset ----------

print("Loading dataset from Hugging Face...")

dataset_name = "camel-ai/physics"

dataset = load_dataset(dataset_name, split="train")  # Adjust split if needed
 
# Convert Hugging Face dataset to instruction fine-tuning format

huggingface_data = []

for row in dataset:

    prompt = f"Solve the following physics problem related to {row['topic;']}, ({row['sub_topic']}): {row['message_1']}"

    response = f"The solution is as follows: {row['message_2']}"

    huggingface_data.append({"prompt": prompt, "response": response})
 
print(f"Converted {len(huggingface_data)} records from Hugging Face dataset.")
 
# ---------- STEP 2: Save as JSON ----------

output_file = "Instruction_fine_tune_dataset_camel_ai_physics.json"

with open(output_file, "w", encoding="utf-8") as f:

    for entry in huggingface_data:

        json.dump(entry, f)

        f.write("\n")
 
print(f"Final instruction fine-tuning dataset saved as: {output_file}")

 
```
